```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#read csv file
s=pd.read_csv("country_vaccinations.csv")
print(s.to_string())
```

```python
#Analyzing data
#head() bydefualt show only 1st 5 row
print(s.head()) 
```

```python
#info() tell us how many non-null value present each column
print(s.info())
```

```python
#data cleaning
#cleaning empty cell
#dropna() used for remove  row that contain empty cell
a=s.dropna()
print(a)
```

```python
#removing duplication
print(s.drop_duplicates())
```

```python
#show blooen value for duplicates data
print(s.duplicated())
```

```python
#show in diagram
s.plot()
plt.show()
```

```python
#scatter plot
s.plot(kind='scatter',x='total_vaccinations',y='daily_vaccinations_per_million')
plt.show()
```

```python
s.plot(kind='hist',x='total_vaccinations',y='daily_vaccinations_per_million')
plt.show()
```
